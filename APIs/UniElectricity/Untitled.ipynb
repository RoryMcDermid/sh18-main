```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import matplotlib.dates

energy_file = pd.read_excel('UniElec092022.xlsx', sheet_name = 1, usecols="B:C")

energy_list = energy_file.values.tolist()

x = energy_list[:30]

x_np = np.array(x)

x_np_time = x_np[:30, 0]
x_list = []

for time in x_np_time:
    x_list.append(time.date())

y_np_cost = x_np[:30, 1]

fig, ax = plt.subplots()

plt.xlabel("Date")
plt.ylabel("Cost")
plt.xticks()

ax.plot(x_list, y_np_cost, linewidth = 2.0)
ax.tick_params(axis='x', labelrotation = 60)
plt.title("Cost of energy: September 2022")
plt.show()



```
